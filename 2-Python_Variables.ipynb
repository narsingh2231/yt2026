### Rules for Naming Variables
- Variable names must adhere to specific rules to prevent syntax errors: 
- Must start with a letter (a-z, A-Z) or an underscore (_).
- Cannot start with a number (0-9).
- Can only contain alpha-numeric characters and underscores (A-z, 0-9, and _).
- Cannot be Python reserved keywords (e.g., if, for, class, print).
- Are case-sensitive (e.g., age, Age, and AGE are three different variables).

### Best Practices - [(PEP 8)](https://peps.python.org/pep-0008/)

 According to the official Style Guide for Python Code (PEP 8), it is recommended to use: 
- Meaningful names to enhance code readability (e.g., user_age is better than ua).
- snake_case (lowercase words separated by underscores) for variable names (e.g., total_count)
