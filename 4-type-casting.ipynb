## Type Casting

Type casting in Python, also known as type conversion, is the process of converting a value from one data type to another.

```python
5 + 5
```

```python
"5" + "5"
```

### Implicit Type Conversion

- This is where the Python interpreter automatically converts one data type to another without any user intervention, typically to prevent data loss. 
- Python safely promotes the "smaller" type to the "larger" type (e.g., an integer to a float)

```python
x = 5
print(type(x))
```

```python
x = 5.0
print(type(x))
```

```python
x = 5.0
x = int(x)
print(type(x))
```

```python
print(x)
```

```python
z = 3.14
z = int(z)
print(z)
```

```python
a= 3.6 # float
b= 3  # int

print(type(a+b))
```

```python
z= float(z)
print(z)
```

```python
x = "5"
z = 3

print(int(x)+z)
```

```python
print(x+str(z))
```

### Explicit Type Conversion

When requires `manual` conversion of a value to a desired data type using built-in functions


Common built-in functions for explicit casting include:
    - `int(x)`: Converts x to an integer. It truncates the decimal part of a float and expects a numeric string.
    - `float(x)`: Converts x to a floating-point number. Useful for precision in calculations.
    - `str(x)`: Converts x to a string representation. Essential for concatenating numbers with text.
    - `bool(x)`: Converts x to a Boolean value (True or False). Most values are True, while 0, None, or empty sequences (like "", [], {}) are False.
    - `list(x)`, `tuple(x)`, `set(x)`, `dict(x)`: Converts sequences or iterables into the respective complex data types.
